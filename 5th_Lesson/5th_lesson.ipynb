```python
def applyAZN(col):
    lst = col.split()
    if 'AZN' in col:
        return float(lst[0])
    if '$' in col:
        return float(lst[0]) * 1.7
    return 0
```

```python
"10500 AZN".split()
```

```python
df2['Qiymet'] = df2['Qiymet'].apply(applyAZN)
```

```python
df2
```

```python
def remove_km(col):
    col = col.strip()
    col = col[:-2]
    
    col = col.replace(' ', '')
    
    return int(col)
```

```python
remove_km('350 000 km')
```

```python
df2['Yurush2'] = df2['Yurush'].apply(remove_km)
df2.drop('Yurush', axis=1, inplace = True)
```
