```python
def change_to_azn(value):
    value = value.split()
    if '$' in value:
        return int(value[0]) * 1.7
    return int(value[0])
```

```python
change_to_azn('10 $')
```

```python
df2 = df[['Buraxilish ili', 'Yanacaq novu', 'Yurush', 'Qiymet']].copy()
```

```python
df2
```

```python
df2['Qiymet'] = df2['Qiymet'].apply(change_to_azn)
```

```python
df2
```

```python
def remove_km(col):
    col = col.strip()
    col = col[:-2]
    col = col.replace(' ', '')
    return int(col)
```

```python
df2['Yurush2'] = df2['Yurush'].apply(remove_km)
```
