```python
data.InvoiceDate.max()
```

```python
temp = data.groupby('CustomerID').max()[['InvoiceDate']]
```

```python
temp
```

```python
temp['Days'] = data.InvoiceDate.max() - temp.InvoiceDate
```

```python
def extract_days(col):
    try:
        col = str(col).split()
        return int(col[0])
    except:
        return 0
temp.Days = temp.Days.apply(extract_days)
```

```python
temp
```

```python
recency = temp[['Days']]
recency
```

```python
def label(col, name, threshold):
    q1 = result[result[name] > result[name].quantile(threshold)][name].quantile(0.25)
    q2 = result[result[name] > result[name].quantile(threshold)][name].quantile(0.50)
    q3 = result[result[name] > result[name].quantile(threshold)][name].quantile(0.75)
    
    if col <= q1:
        return "Worst"
    if col <= q2:
        return "Bad"
    if col <= q3:
        return "Good"
    return "Best"
result['freqLabel'] = result['frequency'].apply(label, args = ('frequency', 0.90))
    
```

```python
def label_recency(col):
    q1 = result['recency'].quantile(0.25)
    q2 = result['recency'].quantile(0.50)
    q3 = result['recency'].quantile(0.75)
    
    if col <= q1:
        return "Best"
    if col <= q2:
        return "Good"
    if col <= q3:
        return "Bad"
    return "Worst"
```

```python
result['recLabel'] = result['recency'].apply(label_recency)
```

```python
result[result.monetary > result.monetary.quantile(0.90)].describe()
```

```python
result['monLabel'] = result['monetary'].apply(label, args = ('monetary', 0.90))
```

```python
def label_frequency(col):
    q1 = result[result.frequency > result.frequency.quantile(0.90)]['frequency'].quantile(0.25)
    q2 = result[result.frequency > result.frequency.quantile(0.90)]['frequency'].quantile(0.50)
    q3 = result[result.frequency > result.frequency.quantile(0.90)]['frequency'].quantile(0.75)
    
    if col <= q1:
        return "Worst"
    if col <= q2:
        return "Bad"
    if col <= q3:
        return "Good"
    return "Best"    
```
