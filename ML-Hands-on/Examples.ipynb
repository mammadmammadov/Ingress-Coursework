```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression

data_root = "https://github.com/ageron/data/raw/main/lifesat/lifesat.csv"
lifesat = pd.read_csv(data_root)
X = lifesat[['GDP per capita (USD)']].values
Y = lifesat[['Life satisfaction']].values

plt.scatter(X, Y, label='Data points', color='blue')

model = LinearRegression()
model.fit(X, Y)

beta_0 = model.intercept_[0]
beta_1 = model.coef_[0][0]
print(f"y = {beta_0} + x * {beta_1}")


X_fit = np.linspace(23500, 62500, 2000).reshape(-1, 1)
Y_fit = model.predict(X_fit)

plt.plot(X_fit, Y_fit, color='red', linewidth=2, label="Linear Regression")

plt.grid(True)
plt.xlabel("GDP per capita (USD)")
plt.ylabel("Life satisfaction")
plt.legend()
plt.axis([23500, 62500, 4, 9])

plt.show()


X_new = np.array([[0]])  # Reshape the input to be a 2D array
predicted_life_satisfaction = model.predict(X_new)[0][0]
print(f"Predicted life satisfaction for GDP per capita of 37000: {predicted_life_satisfaction}")
```
